```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,20,1000) #  1000 evenly spaced points bw 0 and 20
x
```

```python
y = np.sin(x) + 0.2*x
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.show()  # required when not in notebook
```

```python
# labels
plt.plot(x,y)
plt.xlabel('input')
plt.ylabel('output')
plt.title('my plot');
```

```python
X = np.random.randn(100,2)
plt.scatter(X[:,0], X[:,1]);  #scatter col 0 as x, vs col 1 as y; scatter is for a pair of RV's
```

```python
X = np.random.randn(200,2)
X[:50] += 3
y = np.zeros(200)
y[:50] = 1
plt.scatter(X[:,0], X[:,1], c=y);  #c is color, takes an int array, which happens to be y here
```

```python
plt.plot(X[:,0],X[:,1])  # as an exercise, looking at plot vs scatterplot
```

```python
X = np.random.randn(10000)
plt.hist(X);   #histogram is for 1 RV
```

```python
plt.hist(X, bins=50);   # specify the bins
```

```python
X = np.random.random(10000)  # uniform distribution
plt.hist(X, bins=50);
```
