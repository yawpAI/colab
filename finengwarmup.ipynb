```python
n = np.random.multivariate_normal(np.array([0,0]), np.array([[1, -0.5],[-0.5,2]]), size=1000)
n
```

```python
plt.scatter(n[:,0],n[:,1]);
```

```python
n0_mean = n[:,0].sum()/len(n)
n0_mean
```

```python
n.mean()
```

```python
n[:,0].mean()
```

```python
n[:,1].mean()
```

```python
n1_mean = n[:,1].sum() / len(n)
n1_mean
```

```python
 cov01 = (n[:,0]-n0_mean).dot(n[:,1]-n1_mean) / (np.sqrt((n[:,0]-n0_mean).dot(n[:,0]-n0_mean)) * np.sqrt((n[:,1]-n1_mean).dot(n[:,1]-n1_mean)))
```
