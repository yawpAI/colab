```python
import numpy as np
import matplotl6ib.pyplot as plt
from scipy.stats import norm
```

```python
x = np.linspace(-6,6,1000)
fx = norm.pdf(x, loc=0, scale=1)
```

```python
plt.plot(x,fx)
```

```python
Fx = norm.cdf(x, loc=0, scale=1)
plt.plot(x,Fx)
```

```python
logfx = norm.logpdf(x, loc=0, scale=1)
plt.plot(x,logfx)
```

```python
im = Image.open('lena.png')
```

```python
gray = np.mean(im, axis=2)
```

```python
plt.imshow(gray)
```

```python
x = np.linspace(-6,6,50)
fx = norm.pdf(x, loc=0, scale=1)
```

```python
filt = np.outer(fx,fx) # gaussian filter is the outer product of the pdf by itself (a 2d Gaussian distribution)
```

```python
plt.imshow(filt, cmap='gray')
```

```python
from scipy.signal import convolve2d
```

```python
out = convolve2d(gray, filt)  # this will blur the image using the 2d Gaussian filter
```

```python
plt.subplot(1,2,1) # 1 by 2 grid and start in position 1
plt.imshow(gray, cmap='gray')
plt.subplot(1,2,2)
plt.imshow(out, cmap='gray')
```

```python
Hx = np.array([[1,0,-1],[2,0,-2],[1,0,-1]])  #Hx and Hy are Sobel filters
Hx
```

```python
Hy = np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
Hy
```

```python
gray_arr = np.array(gray)
Gx = convolve2d(gray_arr, Hx)
Gy = convolve2d(gray_arr, Hy)
G = np.sqrt(Gx**2 + Gy**2)
```

```python
plt.imshow(G, cmap='gray')
```
